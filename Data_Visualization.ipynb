## 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset with correlated features
num_samples = 100
feature_1 = np.random.normal(0, 1, num_samples)
feature_2 = feature_1 + np.random.normal(0, 0.5, num_samples)  # Correlated with feature_1
feature_3 = np.random.normal(0, 1, num_samples)

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create the heatmap using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, cbar_kws={"shrink": .8})

# Add titles and labels
plt.title('Correlation Matrix Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 1. Violin Plot for Student Grades
import numpy as np
import pandas as pd
import plotly.express as px

# Set a random seed for reproducibility
np.random.seed(15)

# Generate the dataset for student grades
data_grades = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_grades = pd.DataFrame(data_grades)

# Create the violin plot
fig_grades = px.violin(df_grades, x='Grade', y='Score', box=True, points='all',
                       title='Distribution of Scores Across Grade Categories',
                       labels={'Score': 'Scores', 'Grade': 'Grade'},
                       color='Grade')

# Show the plot
fig_grades.show()

# 2. Heatmap for Sales Data
# Set a random seed for reproducibility
np.random.seed(20)

# Generate the dataset for sales data
data_sales = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data_sales)

# Pivot the data to create a matrix for the heatmap
sales_pivot = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create the heatmap
fig_sales = px.imshow(sales_pivot,
                       labels=dict(x="Month", y="Day", color="Sales"),
                       x=sales_pivot.columns,
                       y=sales_pivot.index,
                       title='Sales Variation Across Months and Days')

# Show the plot
fig_sales.show()
```

```python
# Code to Generate the 3D Surface Plot
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Update layout
fig.update_layout(title='3D Surface Plot of z = sin(√(x² + y²))',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'
                  ))

# Show the plot
fig.show()
```

## 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x


```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate y values as the sine of x
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer for the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot
show(p)
```

## 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

# Prepare the output
output_notebook()

# Generate random data
np.random.seed(42)
n = 100  # Number of points
x = np.random.rand(n) * 10  # Random x values from 0 to 10
y = np.random.rand(n) * 10  # Random y values from 0 to 10
sizes = np.random.rand(n) * 50  # Random sizes for the markers
colors = np.random.rand(n)  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a Bokeh figure
p = figure(title="Scatter Plot with Random Sizes and Colors", width=800, height=400)

# Add a scatter renderer
p.scatter('x', 'y', size='sizes', color=linear_cmap('colors', 'Viridis256', low=min(colors), high=max(colors)), source=source, alpha=0.6)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Create a histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value', y_axis_label='Frequency',
           height=400, width=600) # Changed plot_height to height and plot_width to width

# Add a quad renderer to represent the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black')

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
import numpy as np
from bokeh.layouts import layout
from bokeh.palettes import Viridis256

# Activate notebook output
output_notebook()

# Provided dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Bokeh setup for heatmap
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1),
           toolbar_location=None, tools='', width=400, height=400)

# Create a color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=np.min(data_heatmap), high=np.max(data_heatmap))

# Add heatmap as image to the plot
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0, 0), title="Value")
p.add_layout(color_bar, 'right')

# Show the heatmap
show(p)
```
